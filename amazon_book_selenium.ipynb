```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import csv

def setup_driver():
    service = Service(ChromeDriverManager().install())
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def search_amazon(driver, query):
    driver.get("https://amazon.com/")
    wait = WebDriverWait(driver, 10)
    search_box = wait.until(EC.presence_of_element_located((By.ID, "twotabsearchtextbox")))
    search_box.send_keys(query)
    search_box.send_keys(Keys.RETURN)

def scrape_books(driver):
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.s-main-slot div.s-result-item")))

    books = driver.find_elements(By.CSS_SELECTOR, "div.s-main-slot div.s-result-item")
    book_data = []

    for book in books:
        try:

            title_element = book.find_element(By.CSS_SELECTOR, "h2.a-size-base-plus.a-spacing-none.a-color-base.a-text-normal")
            title = title_element.text


            try:
                price_element = book.find_element(By.CSS_SELECTOR, "span.a-price-whole")
                price = price_element.text
            except:
                price = "Price is not available"


            try:
                rating_element = book.find_element(By.CSS_SELECTOR, "span.a-icon-alt")
                rating = rating_element.text.split()[0]
            except:
                rating = "There is no rating"

            book_data.append({"Title": title, "Price": price, "Rating": rating})
        except Exception as e:
            print(f"Error in scraping a book: {e}")
    
    return book_data

def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile: #I used newline='' to avoid error caused by $ in the CSV file
        writer = csv.DictWriter(csvfile, fieldnames=["Title", "Price", "Rating"])
        writer.writeheader()
        writer.writerows(data)
    print(f"Data was saved to {filename}")

if __name__ == "__main__":
    driver = setup_driver()
    try:
        search_amazon(driver, "Python books")
        books = scrape_books(driver)
        save_to_csv(books, "amazon_python_books.csv")
        print(f"Successfully Scraped {len(books)} books.")
    finally:
        driver.quit()
```
